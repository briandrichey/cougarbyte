```python
import json
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parse the JSON file
with open("data.json", "r") as f:
    data = json.load(f)

# Extract the x, y, and z values from the JSON data
x = np.array([d["x"] for d in data])
y = np.array([d["y"] for d in data])
z_r = np.array([d["r"] for d in data])
z_g = np.array([d["g"] for d in data])
z_b = np.array([d["b"] for d in data])
z_id = np.array([d["id"] for d in data])
#z_color = np.array(rgb(z_r,z_g,z_b))

# Create a 3D figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Create the scatter plot
ax.scatter(x, y, z_r,c="red")
ax.scatter(x, y, z_g,c="green")
ax.scatter(x, y, z_b,c="blue")
# ax.contourf(x, y, z[:, 1], zdir="z", offset=0, cmap="Greens")
# ax.contourf(x, y, z[:, 2], zdir="z", offset=0, cmap="Blues")

# Set the axis labels and limits
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_xlim(min(x), max(x))
ax.set_ylim(min(y), max(y))
ax.set_zlim(0, 255)

# Show the plot
plt.show()
plt.savefig("rgb_numpy_matplot_lib_3_color_val.jpeg")
```
